# 1. My Lane as an ML Task

## Refresh / Content Opportunity Scoring
This project is framed as a  classification task.

The goal is to classify whether a content page should be flagged for possible human review based on observable search performance signals. Rather than automatically deciding which pages should be changed, the model helps content teams prioritize which pages deserve investigation first.

## 2. Target or proxy
The target is a proxy label indicating whether a page should be considered a potential content improvement opportunity.

Because the dataset does not contain labels showing which pages were actually refreshed or improved, I created the proxy using a defined rule based on observed search performance.

A page is labeled 1 when:

trend_direction is "down", and
avg_position is greater than 10.

All other pages are labeled 0.

This proxy is intended for decision support, not as a guarantee that a page requires changes.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df.head()
```

```python
df["opportunity_proxy"] = (
    (df["trend_direction"] == "down") &
    (df["avg_position"] > 10)
).astype(int)

df["opportunity_proxy"].value_counts()
```

## 3. Success Metric

The success metric for this classification task is ROC-AUC.

ROC-AUC measures how well the model distinguishes between pages that should be flagged for review and those that should not.

A higher ROC-AUC indicates that the model is better at distinguishing between pages that should be flagged for review and those that should not. This makes it appropriate for a decision-support classification problem where the goal is to prioritize pages for human review.

## 4. The Unit of Analysis

The unit of analysis is one content page.

Each row in the dataset represents one anonymized content page together with its search performance, engagement, and ranking metrics.

The model evaluates each page individually to determine whether it should be flagged for review.

## 5. Why ML Beats a Fixed Rule Here

A fixed rule can identify simple cases, but content performance depends on multiple interacting factors such as search demand, ranking position, click-through rate, engagement, and traffic trends.

Machine learning can learn relationships between these signals that are difficult to capture with manually chosen thresholds.

The model supports content teams by prioritizing pages for review while leaving the final decision to human experts.
